We load our best model from the previous experimentation.

```python
from flair.models import TextClassifier

# load best model from previous training run
classifier = TextClassifier.load('flair_models/old_ccs/best-model.pt')
```

As predict Tweets from our Twitter climate protest dataset.

```python
import pandas as pd
# load prediction sample
ccs_df = pd.read_csv("data/ccs-tweets_5-percent-sample.csv.gz")
ccs_df
```

For reduced runtime, let's use a smaller sample

```python
ccs_sample = ccs_df.sample(n=50000) # Sample random rows
```

We predict the labels in batches (chunks) to not run out of memory.

```python
from tqdm.notebook import tqdm
from flair.data import Sentence

def chunker(seq, size):
    return (seq[pos:pos + size] for pos in range(0, len(seq), size))

tweet_texts = ccs_sample['text']
predicted_labels = []
predicted_scores = []

# Predict labels for each tweet
batch_size = 1000
for chunk in tqdm(chunker(tweet_texts, batch_size), total=int(len(tweet_texts) / batch_size)):

    # Create a list of Flair sentence objects from the 'text' column of the DataFrame
    sentences = [Sentence(text) for text in chunk]
    classifier.predict(sentences, label_name="OLD")

    # Extract the predicted labels and scores
    for sentence in sentences:
        label = sentence.get_label("OLD")
        predicted_labels.append(label.value)
        predicted_scores.append(label.score)

# Add the predictions as new columns to the DataFrame
ccs_sample['predicted_label'] = predicted_labels
ccs_sample['predicted_score'] = predicted_scores

# Display or further process the DataFrame with the predictions
ccs_sample
```

## Descriptive Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'ccs_sample' DataFrame is already created and contains 'created_at', 'predicted_label', and 'protest' columns.
# Ensure 'created_at' is a datetime object.
ccs_sample['created_at'] = pd.to_datetime(ccs_sample['created_at'])

# Create 'month' column
ccs_sample['month'] = ccs_sample['created_at'].dt.to_period('M')

# Group by month, predicted_label, and protest, then count
monthly_counts = ccs_sample.groupby(['month', 'predicted_label', 'protest']).size().reset_index(name='counts')

# Convert 'month' to string before plotting to avoid the TypeError
monthly_counts['month'] = monthly_counts['month'].astype(str)
monthly_counts
```

```python
# Create the plot
plt.figure(figsize=(15, 8)) # Adjust figure size as needed

# Use seaborn for faceting and plotting
g = sns.FacetGrid(monthly_counts, col="predicted_label", hue="protest", col_wrap=3, height=4,) # Adjust col_wrap as needed
g.map(plt.plot, "month", "counts")

# Customize plot appearance
g.set_xticklabels(rotation=45)
g.set_axis_labels("Month", "Counts")
g.fig.suptitle("Monthly Counts of Predicted Labels, Faceted by label", y=1.03) # Adjust y position as needed
g.add_legend()

plt.tight_layout() # Improve layout
plt.show()
```

## Explaining observations

We can use regression analysis to explain / predict dependencies between variables. When correlating social media metrics such as likes or replies, we need to take care for the power law distribution of this data. This kind of data can be applied with a poisson regression model.

**Let's look at replies to tweets. We ask: To what extent does the use of offensive language contribute to the increase of likes? **

```python
original_tweets = ccs_sample[ccs_sample['tweet_type'] == "reply"].copy()
original_tweets["is_offensive"] = original_tweets["predicted_label"] == "OFFENSE"
corr_data = original_tweets[["is_offensive", "likes", "retweets", "replies", "quotes"]]
corr_data.shape
```

Run the statistical model

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Assuming 'data' is your DataFrame
corr_data['is_offensive'] = corr_data['is_offensive'].astype(int) # Convert boolean to int (True to 1, False to 0)

# Poisson regression using the formula API
model_poisson = smf.poisson('likes ~ is_offensive', data=corr_data).fit()
print(model_poisson.summary())
```

ChatGPT summarizes the result as follows:

The Poisson regression results indicate a statistically significant positive relationship between `is_offensive` and the number of `likes`. Here’s a breakdown of the key points:

1. **Coefficient for `is_offensive` (0.3377)**: This positive coefficient suggests that tweets classified as offensive are associated with a higher expected count of likes. In Poisson regression, coefficients are in log scale, meaning that each unit increase in `is_offensive` (i.e., moving from `False` to `True`) is associated with an increase in the log count of likes by 0.3377.

2. **Interpretation in Percent Terms**: To understand this effect in percentage terms, we can exponentiate the coefficient. The expected increase in likes for offensive tweets is `exp(0.3377) - 1 ≈ 0.401`, or about **40.1% more likes** compared to non-offensive tweets, holding other factors constant.

3. **Statistical Significance**: The very low p-value (`P>|z| = 0.000`) for `is_offensive` indicates that this effect is statistically significant, meaning there's strong evidence that offensiveness is associated with the number of likes.

4. **Pseudo R-squared (0.0003281)**: This metric is typically low for count models, and a low value is expected. However, it shows that the model explains only a tiny fraction of the variance in likes, indicating that while `is_offensive` has a significant effect, it’s likely not a major driver of likes on its own.

This model implies that offensiveness in tweets is associated with a higher number of likes, but other factors likely play a role in predicting like counts, as indicated by the low pseudo R-squared.
